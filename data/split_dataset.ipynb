```python
import torch
import numpy as np
import os 


def split_dataset(data:torch.Tensor or list,fp,train_set:float=0.7,test_set:float=0.2,val_set:float=0.1):
    length = len(data)
    train,test,val=torch.randperm(length).split([int(train_set*length),int(test_set*length),length-int(train_set*length)-int(test_set*length)])
    index = {}
    index.update({'train':train,'test':test,'val':val})
    torch.save(index,fp)

length =  len(os.listdir('./DSB2018/train'))
data = torch.arange(length)
split_dataset(data,'/root/autodl-tmp/splitted_index/dsb_splitted_indices.pt',train_set=0.7,test_set=0.2,val_set=0.1)
print(length,data)

length =  len(os.listdir('./cpm17/train/Images'))
data = torch.arange(length)
print(length,data)
split_dataset(data,'/root/autodl-tmp/splitted_index/cpm_splitted_indices.pt',train_set=0.9,test_set=0,val_set=0.1)
```

```python
import os
import torch
import matplotlib.pyplot as plt

img_list = os.listdir('./conic/images')
img_list.sort()
label_list = os.listdir('./conic/labels')
label_list.sort()

imgs = torch.zeros(len(img_list),3,256,256)
labels = torch.zeros(len(label_list),2,256,256)
for i in range(len(img_list)):
    img = torch.load(os.path.join('./conic/images',img_list[i]))
    label = torch.load(os.path.join('./conic/labels',label_list[i]))
    imgs[i] = img
    labels[i] = label
torch.save(imgs,'./conic/imgs.pt')
torch.save(labels,'./conic/labels.pt')
```
